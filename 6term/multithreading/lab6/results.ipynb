```python
import matplotlib.pyplot as plt

blocking_times = {
  3: [911, 18171, 63575, 178040],
  4: [771, 12993, 44681, 134263],
  5: [641, 9220, 34038, 112911],
  6: [667, 10759, 36729, 123695]
}

non_blocking_times = {
  3: [776, 17684, 61226, 176121],
  4: [701, 12312, 42851, 121053],
  5: [607, 8730, 32028, 102939],
  6: [627, 10340, 35039, 114286]
}

matrix_sizes = [750, 1500, 2250, 3000]

fig, axs = plt.subplots(4, 1, figsize=(7, 20))

axs[0].plot(matrix_sizes, blocking_times[3], marker='o', label='Блокуючий')
axs[0].plot(matrix_sizes, non_blocking_times[3], marker='o', label='Неблокуючий')
axs[0].set_title('3 процеси')
axs[0].set_xlabel('Розмір матриці, nxn')
axs[0].set_ylabel('Час, мс')
axs[0].legend()

axs[1].plot(matrix_sizes, blocking_times[4], marker='o', label='Блокуючий')
axs[1].plot(matrix_sizes, non_blocking_times[4], marker='o', label='Неблокуючий')
axs[1].set_title('4 процеси')
axs[1].set_xlabel('Розмір матриці, nxn')
axs[1].set_ylabel('Час, мс')
axs[1].legend()

axs[2].plot(matrix_sizes, blocking_times[5], marker='o', label='Блокуючий')
axs[2].plot(matrix_sizes, non_blocking_times[5], marker='o', label='Неблокуючий')
axs[2].set_title('5 процесів')
axs[2].set_xlabel('Розмір матриці, nxn')
axs[2].set_ylabel('Час, мс')
axs[2].legend()

axs[3].plot(matrix_sizes, blocking_times[6], marker='o', label='Блокуючий')
axs[3].plot(matrix_sizes, non_blocking_times[6], marker='o', label='Неблокуючий')
axs[3].set_title('6 процесів')
axs[3].set_xlabel('Розмір матриці, nxn')
axs[3].set_ylabel('Час, мс')
axs[3].legend()

plt.tight_layout()
plt.show()
```
