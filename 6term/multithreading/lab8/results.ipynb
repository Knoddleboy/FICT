```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dim_df = pd.DataFrame({
  'rows': [500, 1000, 1500, 2000, 2500, 3000],
  'client': [165, 626, 2126, 4464, 8455, 14058],
  'server': [116, 571, 2000, 3958, 7971, 13411],
})

plt.plot(dim_df['rows'], dim_df['client'], label='Генерація на клієнті')
plt.plot(dim_df['rows'], dim_df['server'], label='Генерація на сервері')
plt.legend()
plt.ylabel('Час, мс')
plt.xlabel('Розмір матриці, nxn')
plt.show()
```
