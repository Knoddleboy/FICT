```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dim_df = pd.DataFrame({
  'rows': [750, 1500, 2250, 3000],
  '3t': [691, 13609, 45726, 97471],
  '4t': [575, 11178, 41029, 90432],
  '5t': [737, 11216, 41925, 93247],
  '6t': [728, 11851, 46995, 102593],
})
```

```python
plt.plot(dim_df['rows'], dim_df['3t'], label='3 потоки')
plt.plot(dim_df['rows'], dim_df['4t'], label='4 потоки')
plt.plot(dim_df['rows'], dim_df['5t'], label='5 потоки')
plt.plot(dim_df['rows'], dim_df['6t'], label='6 потоки')
plt.legend()
plt.ylabel('Час, мс')
plt.xlabel('Розмір матриці, nxn')
plt.show()
```
