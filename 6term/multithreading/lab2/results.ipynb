```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dim_df = pd.DataFrame({
  'rows': [500, 1000, 1500, 2000, 2500, 3000],
  'block-striped': [1.202, 2.943, 6.015, 9.525, 16.096, 23.482],
  'fox': [0.783, 2.01, 3.78, 6.182, 7.1, 12.116],
  'block-striped_square': [0.133, 1.196, 4.963, 15.287, 36.265, 75.14],
  'fox_square': [0.128, 0.589, 2.608, 8.016, 15.612, 47.429],
})
```

```python
plt.plot(dim_df['rows'], dim_df['block-striped'], label='Block-Striped')
plt.plot(dim_df['rows'], dim_df['fox'], label='Fox')
plt.legend()
plt.ylabel('Time, s')
plt.xlabel('Matrix Size, nx1750')
plt.show()
```

```python
plt.plot(dim_df['rows'], dim_df['block-striped_square'], label='Block-Striped')
plt.plot(dim_df['rows'], dim_df['fox_square'], label='Fox')
plt.legend()
plt.ylabel('Time, s')
plt.xlabel('Matrix Size, nxn')
plt.show()
```
