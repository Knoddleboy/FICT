```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dim_df = pd.DataFrame({
  'rows': [750, 1500, 2250, 3000],
  '2t': [1.02, 1.56, 3.70, 4.90],
  '3t': [1.77, 2.56, 5.98, 6.88],
  '4t': [1.48, 2.03, 4.94, 6.21],
})
```

```python
plt.plot(dim_df['rows'], dim_df['2t'], label='2 потоки')
plt.plot(dim_df['rows'], dim_df['3t'], label='3 потоки')
plt.plot(dim_df['rows'], dim_df['4t'], label='4 потоки')
plt.legend()
plt.ylabel('Прискорення')
plt.xlabel('Розмір матриці, nxn')
plt.show()
```

```python
t2_df = pd.DataFrame({
  'rows': [750, 1500, 2250, 3000],
  'es': [514, 6260, 51636, 155338],
  'fjp': [505, 4019, 13972, 31719],
})

t3_df = pd.DataFrame({
  'rows': [750, 1500, 2250, 3000],
  'es': [534, 6334, 52649, 134885],
  'fjp': [301, 2479, 8806, 19616],
})

t4_df = pd.DataFrame({
  'rows': [750, 1500, 2250, 3000],
  'es': [264, 3821, 28854, 79311],
  'fjp': [178, 1882, 5842, 12762],
})
```
