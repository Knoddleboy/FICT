```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Data2.csv', on_bad_lines='skip', delimiter=';', encoding='cp1252')
df
```

```python
df.dtypes
```

```python
df.columns = ['country_name', 'region', 'gdp_per_capita', 'population', 'co2_emission', 'area']
df
```

```python
df.set_index('country_name', inplace=True)
df
```

```python
df.gdp_per_capita = df.gdp_per_capita.astype(str).str.replace(',', '.').astype(float)
df.co2_emission = df.co2_emission.astype(str).str.replace(',', '.').astype(float)
df.area = df.area.astype(str).str.replace(',', '.').astype(float)
df
```

```python
df.dtypes
```

```python
df.gdp_per_capita = df.gdp_per_capita.abs()
df.area = df.area.abs()
df
```

```python
df.isna().any()
```

```python
df.gdp_per_capita.fillna(df.gdp_per_capita.mean(), inplace=True)
df.population.fillna(df.population.mean(), inplace=True)
df.co2_emission.fillna(df.co2_emission.mean(), inplace=True)
df
```

```python
df.isna().any()
```

```python
df.population = df.population.astype(int)
df
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 14))
axes[0, 0].boxplot(df.gdp_per_capita)
axes[0, 0].set_title('GDP per capita')

axes[0, 1].boxplot(df.population)
axes[0, 1].set_title('Population')

axes[1, 0].boxplot(df.co2_emission)
axes[1, 0].set_title('CO2 emissions')

axes[1, 1].boxplot(df.area)
axes[1, 1].set_title('Area')
```

```python
df.hist(figsize=(20, 16))
```

```python
df['density'] = df.population / df.area
df
```

```python
# 2.1 Яка країна має найбільший ВВП на людину (GDP per capita)?
df.loc[[df.gdp_per_capita.idxmax()]]
```

```python
# 2.2 Яка має найменшу площу?
df.loc[[df.area.idxmin()]]
```

```python
# 3. В якому регіоні середня площа країни найбільша?
gr_region = df.groupby([df.region])

group = gr_region.mean()
group.loc[[group.area.idxmax()]]
```

```python
# 4.1 Знайдіть країну з найбільшою щільністю населення у світі?
df.loc[[df.density.idxmax()]]
```

```python
# 4.2 У Європі та центральній Азії?
df.loc[[
    df[df.region == 'Europe & Central Asia'].density.idxmax()
]]
```

```python
# 5. Чи співпадає в якомусь регіоні середнє та медіана ВВП?
df['gdp'] = df.population * df.gdp_per_capita
df
```

```python
agg_gdp = gr_region.gdp.aggregate(['mean', 'median'])
agg_gdp
```

```python
agg_gdp[agg_gdp['mean'] == agg_gdp['median']]
```

```python
# ВВП не співпадає. Знайдемо найближчі до співпадіння середнє та медіану ВВП
agg_gdp['difference'] = (agg_gdp['mean'] - agg_gdp['median']).abs()
agg_gdp
```

```python
agg_gdp.loc[[agg_gdp.difference.idxmin()]]
```

```python
# 6. Вивести топ 5 країн та 5 останніх країн по ВВП та кількості СО2 на душу населення.
df['co2_emission_per_capita'] = df.co2_emission / df.population
df
```
